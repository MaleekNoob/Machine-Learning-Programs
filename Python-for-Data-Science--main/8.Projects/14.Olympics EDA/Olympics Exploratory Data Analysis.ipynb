### Total Number of women athletes taken part in each olympic

```python
female_participate = data[(data.Sex=='F')& (data.Season=='Summer')][['Sex','Year']]
female_participate = female_participate.groupby('Year').count().reset_index()
female_participate
```

```python
women_olym = data[(data.Sex=='F')& (data.Season =='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data = women_olym,palette="Spectral")
plt.title("Women Participation")
```

```python
part = women_olym.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of female athletes over time')
```

## Gold Medal Athletes

```python
goldmedals = data[(data.Medal=='Gold')]
```

```python
goldmedals.head()
```

```python
# take only values that are different from NaN
gold_medals = goldmedals[np.isfinite(goldmedals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age']>60].count()
```

```python
sport_event = gold_medals["Sport"][gold_medals['Age']>60]
```

```python
sport_event
```

```python
plt.figure(figsize=(12,6))
plt.tight_layout()
sns.countplot(sport_event)
```

```python
gold_medals.Region.value_counts().reset_index()
```

```python
totalGoldMedals = gold_medals.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x="index",y="Medal",data= totalGoldMedals,
               height =5,kind="bar",palette="rocket")

g.despine(left=True)
plt.title("Gold Medals Per Country")
```
