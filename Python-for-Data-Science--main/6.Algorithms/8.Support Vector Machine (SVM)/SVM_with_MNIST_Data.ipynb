# Support Vector Machine (SVM)

## What is Support Vector Machine?
The objective of the support vector machine algorithm is to find a hyperplane in an N-dimensional space(N â€” the number of features) that distinctly classifies the data points.



```python
from IPython.display import Image
Image('1.png')
```

To separate the two classes of data points, there are many possible hyperplanes that could be chosen. Our objective is to find a plane that has the maximum margin, i.e the maximum distance between data points of both classes. Maximizing the margin distance provides some reinforcement so that future data points can be classified with more confidence

## Hyper-planes and Support Vectors

```python
Image('2.png')
```

Hyperplanes are decision boundaries that help classify the data points. Data points falling on either side of the hyperplane can be attributed to different classes. Also, the dimension of the hyperplane depends upon the number of features. If the number of input features is 2, then the hyperplane is just a line. If the number of input features is 3, then the hyperplane becomes a two-dimensional plane. It becomes difficult to imagine when the number of features exceeds 3.

```python
Image('3.png')
```

Support vectors are data points that are closer to the hyperplane and influence the position and orientation of the hyperplane. Using these support vectors, we maximize the margin of the classifier. Deleting the support vectors will change the position of the hyperplane. These are the points that help us build our SVM.

## Is Small Margin is Good or Large Margin is Good?
In general, a larger margin is considered desirable in SVMs. A larger margin indicates a greater degree of separation between the classes, which can lead to better generalization and improved performance on unseen data. A wider margin allows for a greater tolerance to errors and noise in the training data, reducing the risk of overfitting.

A small margin, on the other hand, may indicate that the decision boundary is closer to the training examples, potentially leading to overfitting and reduced generalization ability. It may also be more sensitive to noise and outliers in the data, increasing the risk of misclassification.

## Nonlinear SVMs and the Kernel Trick
In real-world scenarios, data is rarely linearly separable. To handle such cases, Nonlinear SVMs come into play. These SVMs allow for more complex decision boundaries by utilizing the "Kernel Trick." The Kernel Trick transforms the original input space into a higher-dimensional feature space, where data points may become linearly separable.

The key idea is to implicitly map the input data into a higher-dimensional space without explicitly computing the coordinates of the data in that space. This allows Nonlinear SVMs to learn and identify nonlinear patterns in the data.

## Popular Kernels for Nonlinear SVMs
Kernels are functions that measure the similarity between pairs of data points. They play a crucial role in Nonlinear SVMs. Here are some commonly used kernels:

1. Polynomial Kernel: This kernel computes the similarity between two vectors by raising their dot product to a user-specified power. It is effective in capturing polynomial relationships between features.

2. Radial Basis Function (RBF) Kernel: The RBF kernel computes the similarity between two vectors based on the Gaussian distribution. It is particularly useful in capturing complex and non-linear patterns in the data.

3. Sigmoid Kernel: The sigmoid kernel measures similarity based on the hyperbolic tangent function. It can be useful for capturing certain types of nonlinear relationships.

### Advantages of SVMs:

Effective in High-Dimensional Spaces: 

* SVMs perform well even when the number of dimensions is greater than the number of samples, making them suitable for applications with many features.

* Robust to Outliers: SVMs are less affected by outliers since they focus on the support vectors that lie closest to the decision boundary.

* Memory Efficient: SVMs only rely on the support vectors for decision-making, making them memory-efficient for large datasets.

## Polynomial Kernel in SVM

In Support Vector Machines (SVMs), the polynomial kernel is a popular choice for nonlinear classification tasks. It is a mathematical function that measures the similarity between pairs of data points in a higher-dimensional space. The polynomial kernel allows SVMs to handle nonlinear relationships between features by implicitly mapping the data into a higher-dimensional feature space.

The polynomial kernel takes the dot product of the two input vectors and raises it to the power of d. The resulting value represents the similarity between the vectors in the higher-dimensional feature space. The degree parameter (d) controls the complexity of the decision boundary. Higher values of d allow for more complex decision boundaries that can capture intricate nonlinear patterns in the data.

By using the polynomial kernel, SVMs can find nonlinear decision boundaries that separate different classes of data. This enables SVMs to handle datasets that are not linearly separable in the original input space.

### Advantages of Polynomial Kernel SVM
Polynomial kernel SVM has several advantages over other machine learning algorithms:

1. It is effective for high-dimensional datasets with a relatively small number of training samples.
2. It can capture non-linear relationships between the input data.
3. It is computationally efficient, particularly for low-degree polynomial functions.
4. It can be combined with other kernel functions to handle more complex datasets.

### Applications of Polynomial Kernel SVM
Polynomial kernel SVM has a wide range of applications in various fields, including:

### Image Classification

Polynomial kernel SVM can be used to classify images based on their features. For example, it can be used to classify images of handwritten digits based on their pixel values.

#### Text classification

Polynomial kernel SVM can be used to classify text data, such as emails or social media posts, based on their content.

#### Bioinformatics

Polynomial kernel SVM can be used to classify DNA sequences based on their features, such as nucleotide composition or secondary structure.

### Choosing the Right Parameters for Polynomial Kernel SVM
The performance of polynomial kernel SVM depends on the choice of its parameters, such as the degree of the polynomial function and the regularization parameter. Choosing the right parameters is crucial for achieving good performance on the test data. There are several techniques for choosing the parameters, such as grid search and cross-validation.

# Radial Basis Function Kernel
In Support Vector Machines (SVMs), the Radial Basis Function (RBF) kernel, also known as the Gaussian kernel, is a widely used choice for nonlinear classification tasks. It allows SVMs to handle complex and nonlinear relationships between features by implicitly mapping the data into a higher-dimensional feature space.

```python
Image('5.png')
```

The RBF kernel allows SVMs to capture complex and nonlinear patterns in the data. It is particularly effective in cases where the decision boundary is irregular or when there are overlapping classes. By mapping the data into a higher-dimensional feature space, the RBF kernel enables SVMs to create flexible and adaptable decision boundaries.

When using the RBF kernel, it is important to choose an appropriate value for the Î³ parameter. If `Î³` is too large, the RBF kernel can create decision boundaries that overfit the training data and perform poorly on unseen data. If `Î³` is too small, the decision boundaries may become overly smoothed and may not capture the intricacies of the data.

Selecting the optimal value for `Î³` often involves hyperparameter tuning using techniques like cross-validation to find the best trade-off between model complexity and generalization.

The Radial Basis Function (RBF) kernel in Support Vector Machines (SVMs) offers several advantages that make it a popular choice for nonlinear classification tasks. Let's explore some of the advantages of using the RBF kernel in SVMs:

1. <b>Ability to Capture Complex Patterns:</b> The RBF kernel is capable of capturing complex and nonlinear relationships between features in the data. It allows SVMs to create decision boundaries that can flexibly adapt to intricate patterns and variations in the data. This makes the RBF kernel suitable for datasets with complex structures and overlapping classes.

2. <b>Flexibility in Decision Boundary Shapes:</b> The RBF kernel provides more flexibility in shaping the decision boundaries compared to linear kernels. It can generate decision boundaries that are curved, irregular, or contain non-convex regions. This versatility allows SVMs with the RBF kernel to handle a wide range of classification problems, including those with complex data distributions.

3. <b>Effective Handling of High-Dimensional Data:</b> SVMs with the RBF kernel can effectively handle high-dimensional data, where the number of features is large. The RBF kernel implicitly maps the data to a higher-dimensional feature space, making it possible to separate classes that may be inseparable in the original input space. This capability makes the RBF kernel well-suited for applications with many features, such as text classification, image recognition, and genomics.

4. <b>Robustness to Outliers:</b> The RBF kernel is less sensitive to outliers compared to some other kernels, such as the polynomial kernel. The influence of outliers on the decision boundary is typically limited to the support vectors that are closest to the boundary. This robustness helps SVMs with the RBF kernel to focus on the most relevant data points and avoid being excessively influenced by noisy or outlier samples.

5. <b>Generalization to Unseen Data:</b> The RBF kernel, when properly tuned, can provide good generalization performance. It can capture the underlying distribution of the data and effectively classify unseen samples. By adapting the decision boundary to the training data, SVMs with the RBF kernel have the potential to generalize well and make accurate predictions on new, unseen data.

6. <b>Efficient Computation:</b> Despite operating in a higher-dimensional feature space, the RBF kernel can be computed efficiently using techniques like the kernel trick. This allows SVMs with the RBF kernel to handle large-scale datasets without incurring excessive computational costs.

```python
def view_digit(index):
    plt.imshow(digits.images[index] , cmap = plt.cm.gray_r)
    plt.title('Orignal it is: '+ str(digits.target[index]))
    plt.show()
```

```python
view_digit(4)
```

```python
main_data = digits['data']
targets = digits['target']
```

```python
from sklearn import svm

svc = svm.SVC(gamma=0.001 , C = 100)    
# GAMMA is a parameter for non linear hyperplanes. 
# The higher the gamma value it tries to exactly fit the training data set
# C is the penalty parameter of the error term. 
# It controls the trade off between smooth decision boundary and classifying the training points correctly.

svc.fit(main_data[:1500] , targets[:1500])   

predictions = svc.predict(main_data[1501:])  

list(zip(predictions , targets[1501:]))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(predictions, targets[1501:])

conf_matrix = pd.DataFrame(data = cm)

plt.figure(figsize = (8,5))

sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu");
```

```python
from sklearn.metrics import classification_report

print(classification_report(predictions, targets[1501:]))
```

```python
preprocessed_image = digits.images[4] # Preprocess the image in the same way as training data

# Reshape the preprocessed image to match the shape expected by the SVM model
reshaped_image = preprocessed_image.reshape(1, -1)  # Assuming you flattened the image

# Make predictions on the reshaped image
prediction = svc.predict(reshaped_image)

print("Predicted output:", prediction)
```
