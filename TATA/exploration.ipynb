```python
import pandas as pd

data = pd.read_excel('Online Retail.xlsx')

print(data.head(10))
```

```python
data.info()
```

```python
data_sample = data.sample(n=20, random_state=42)
data_sample = data_sample[['Description', 'Quantity', 'UnitPrice', 'Country']]
data_sample['TotalPrice'] = data_sample['Quantity'] * data_sample['UnitPrice']

print(data_sample)
```

```python
import matplotlib.pyplot as plt

# plot the distribution of Quantity
plt.hist(data['Quantity'], bins=150, range=(0, 100))
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# plot the distribution of UnitPrice
plt.hist(data['UnitPrice'], bins=150, range=(0, 100))
plt.title('Distribution of UnitPrice')
plt.xlabel('UnitPrice')
plt.ylabel('Frequency')
plt.show()

# plot the distribution of CustomerID
plt.hist(data_sample['TotalPrice'], bins=150, range=(0, 100))
plt.title('Distribution of TotalPrice')
plt.xlabel('TotalPrice')
plt.ylabel('Frequency')
plt.show()
```
